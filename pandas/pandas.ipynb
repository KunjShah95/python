```python
pivot_table = df.pivot_table(index='season', columns='team1', values='result_margin', aggfunc='sum')
pivot_table.head() # used to display the first few rows of the pivot table

pivot_table.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.ylabel('Total Result Margin (Runs)')
plt.title('Total Result Margin by Team1 per Season')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
